# 2. Создадим пропуски в данных и заполним их после

```python
import random
from numpy import NaN

rnd = random.Random()

nan_data = df.copy()
for i in range(nan_data.shape[0]): #возвращает размерность таблицы в виде кортежа, 0 - вернет количество строк
    if rnd.randint(0, 100) < 10:
        select_column = rnd.choice(nan_data.columns) #дает случайное значение из представленных
        print(select_column, i)
        nan_data[select_column][i] = NaN
```

```python
print('\nИзначально')
print(df.dropna().count())  #dropna - удаляет пропуски
print('\nПосле обработки')
print(nan_data.dropna().count())
nan_data.head(55)
```

```python
const_data = nan_data.fillna(0) #заполнение константой
const_data.head(50)
```

```python
mean_data = nan_data.fillna(nan_data.mean()) #взятие среднего по столбцам
print(f'Mean: \n{nan_data.mean()}')
mean_data.head(50)
```

```python
nan_data_fill_near_value = nan_data.fillna(method='pad') #заполнение через соседа
nan_data_fill_near_value.head(50)
```

```python
nan_data_interpolate = nan_data.interpolate(method='cubic') #интерполяция столбцов с заполнением пропусков
nan_data_interpolate.head(50)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline
```

```python
ages = np.unique(df['age'])
colors = [plt.cm.tab10(i/float(len(ages)-1)) for i in range(len(ages))]
print(colors)
```

```python
plt.figure(figsize=(16, 12), dpi= 80, facecolor='w', edgecolor='k')

for i, age in enumerate(ages):
    plt.scatter('chol', 'age', 
                data=df.loc[df.age==age, :], 
                s=20, label=age)
plt.gca().set(xlim=(100,400 ), ylim=(0, 100),
              xlabel='chol', ylabel='age')
plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Scatterplot of Age vs Chol", fontsize=22)
plt.legend(fontsize=10.5)
plt.show() 
```

```python
df_counts = df.groupby(['trestbps','chol']).size().reset_index(name='counts')
fig, ax = plt.subplots(figsize=(16,10), dpi = 80)
sns.scatterplot(df_counts.trestbps, df_counts.chol, size=df_counts.counts,ax=ax)
plt.title('trestbps & chol', fontsize=22)
plt.show() #график counts plot - чем больше точка, тем больше концентрация точек вокруг нее
```

```python
new_df = df.groupby('sex').size()
print(new_df)
plot = new_df.plot.pie(x='sex', figsize=(5, 5), labels=['female','male'], label='sex',shadow=True ) #диаграмма распределения заболевания в зависимости от пола
```
