# 2. Создадим пропуски в данных и заполним их после

```python
import random
from numpy import NaN

rnd = random.Random()

nan_data = df.copy()
for i in range(nan_data.shape[0]): #возвращает размерность таблицы в виде кортежа, 0 - вернет количество строк
    if rnd.randint(0, 100) < 10:
        select_column = rnd.choice(nan_data.columns) #дает случайное значение из представленных
        print(select_column, i)
        nan_data[select_column][i] = NaN
```

Для создания пропусков, просто возьмем случайные значения столбцов и строк и заменим содержимое в ячейках таблицы на NaN. Далее выведем списком название колонки и номер строки в которых происхлдит замена.

```python
print('\nИзначально')
print(df.dropna().count())  #dropna - удаляет пропуски
print('\nПосле обработки')
print(nan_data.dropna().count())
nan_data.head(55)
```

Посмотрим на количество элементов до и после обработки. В первом случае у нас выдано количество без пропусков изначального набора данных. Во втором - после удаления строк, в которых есть хотя-бы один NaN

```python
const_data = nan_data.fillna(0) #заполнение константой
const_data.head(50)
```

Построим различные графики и посмотрим, что они могут нам показать. В начале произведем настройку и подключение различных библиотек

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline

df['sex']=df['sex'].replace(to_replace =0,value='female')
df['sex']=df['sex'].replace(to_replace =1,value='male')
```

Построим самую простую диаграмму, которая показывает соотношение полов в таблице данных. Для этого переопренделим данные по полу 1- male 0 -female для удобства

Как видно из графика мужская половина преобладает на тестовом наборе данных

Посмотрим countsplot - график показывающий толщиной точки большую концентрацию элементов. Посмотрим как зависят удары сердца в секудну от холестерина. 

```python
df_counts = df.groupby(['trestbps','chol']).size().reset_index(name='counts')
fig, ax = plt.subplots(figsize=(16,10), dpi = 80)
sns.scatterplot(df_counts.trestbps, df_counts.chol, size=df_counts.counts,ax=ax)
plt.title('trestbps & chol', fontsize=22)
plt.show() #график counts plot - чем больше точка, тем больше концентрация точек вокруг нее
```

Оказывается, что никак не зависит. Однако можно заметить интересную особенность: Значения ударов в секунду чаще всего кратны 10, вероятно измерения округлялись до кратных значений.

Далее посмотрим на зависимость холестерина от возраста. Также окрасим точку в зависимости от пола. Уберем предупреждения, так как они не красивые. Зная, что норма для человека 200 mg/dL, проведем анализ.

Рассмотрим функцию pairplot которая строит парные графики в зависимости от переданных столбцов исходного набора данных. Выберем данные для вывода: Холестерин, возраст и артериальное давление. Столбиковые диаграммы показывают простое распределение по количеству холестерина/возраста/артериальному давлению. Другие-же диаграммы читать неудобно из-за большого количества точек, лежащих в одном месте.

```python
sns.distplot(df.trestbps)
#Артериальное давление в покое в мм.рт.ст норма - 120-130
```

Здесь мы можем заметить что среднее значение около 125-130 мм.рт.столба. что соответствует среднему значению здорового человека. Однако максимальное количество человек с наиболее частым значением лежит чуть выше среднего значения, что говорит о проблемах со здоровьем или о возможных ошибках в измерении

```python
g = sns.jointplot(x="oldpeak", y="thalach", data=df, kind="reg")
#Депрессия, вызванная физическими упражнениями относительно отдыха 6 - сильная усталость
#достигнут максимальный сердечный ритм от 130-160 норма для здорового человека
```

Чем выше максимальный сердечный ритм, тем сложнее справиться сердцу с нагрузкой. Поэтому высоких значений депрессии почти нет, а значения ритма больше в средней зоне, однако среднее значение испытуемых выше верхней границы. 

```python
sns.boxplot(y="ca", x="age", data=df, orient="h")
#ca - канамбер крупных сосудов (0-4), окрашенный с помощью флуороскопии... че?
```

Коробка показывает интерквартильный размах распределения, то есть соответственно 25% (Q1) и 75% (Q3) перцентили. Черта внутри коробки обозначает медиану распределения.
Усы отображают весь разброс точек кроме выбросов, то есть минимальные и максимальные значения, которые попадают в промежуток (Q1 - 1.5*IQR, Q3 + 1.5*IQR), где IQR = Q3 - Q1 — интерквартильный размах. Точками на графике обозначаются выбросы (outliers) — те значения, которые не вписываются в промежуток значений, заданный усами графика.

График показывает, что канамберы крупных сосудов 0 типа (здоровые) встречаются у большей части испытуемых. в то время как осложнения появляются у испытуемых с более почтенным возрастом. так канамберы 3 типа встречаются у испытуемых с 45 до 75 лет, а среднее значение возраста почти 60.
